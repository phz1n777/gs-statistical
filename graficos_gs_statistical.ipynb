```python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
anos = ['2022', '2023', '2024']
hidraulica_percent = [60.5, 50.16, 46.86]
eolica_percent = [12.8, 13.19, 14]
solar_percent = [3.8, 5.19, 6.97]
termica_percent = [21.6, 18.27, 16.12]
nuclear_percent = [1.1, 0.92, 0.86]
mmgd_percent = [0, 12.27, 15.18]  # MMGD é 0% em 2022 e com valores a partir de 2023

# Posicionamento das barras
x = np.arange(len(anos))

# Largura das barras
largura = 0.15

# Criando o gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Adicionando as barras para cada fonte de energia
ax.bar(x - 2*largura, hidraulica_percent, largura, label='Hidráulica', color='blue')
ax.bar(x - largura, eolica_percent, largura, label='Eólica', color='green')
ax.bar(x, solar_percent, largura, label='Solar', color='orange')
ax.bar(x + largura, termica_percent, largura, label='Térmica', color='red')
ax.bar(x + 2*largura, nuclear_percent, largura, label='Nuclear', color='purple')
ax.bar(x + 3*largura, mmgd_percent, largura, label='MMGD', color='brown')

# Definindo rótulos e título
ax.set_xlabel('Ano')
ax.set_ylabel('Participação Percentual')
ax.set_title('Proporção de Produção por Ano (2022-2024)')
ax.set_xticks(x)
ax.set_xticklabels(anos)

# Adicionando a legenda
ax.legend()

# Adicionando as linhas de fundo (grid)
ax.grid(True, axis='y', linestyle='--', linewidth=0.7, color='gray', alpha=0.5)

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de produção em MW para cada fonte de energia em 2022, 2023 e 2024
producoes = {
    'Hidráulica': [109189, 108392, 108001],
    'Eólica': [23162, 28506, 32264],
    'Solar': [6917, 11212, 16075],
    'Térmica': [39090, 39483, 37163],
    'Nuclear': [1990, 1990, 1990],
    'MMGD': [0, 26510, 34988]
}

# Preparando os dados para o gráfico
anos = ['2022', '2023', '2024']
labels = list(producoes.keys())  # Tipos de energia

# Convertendo as listas de produção em um formato adequado para empilhar
dados = np.array([producoes[label] for label in labels])

# Criando o gráfico de barras empilhadas
fig, ax = plt.subplots(figsize=(10, 6))

# Plotando as barras empilhadas
ax.bar(anos, dados[0], label=labels[0], color='#87CEEB')  # Cor da Hidráulica
for i in range(1, len(dados)):
    ax.bar(anos, dados[i], bottom=np.sum(dados[:i], axis=0), label=labels[i])

# Adicionando título e rótulos
ax.set_title('Distribuição da Produção de Energia por Fonte (2022-2024)', fontsize=16)
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Produção Total de Energia (MW)', fontsize=14)

# Adicionando a legenda
ax.legend(title='Fontes de Energia', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustando o layout
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos (incluindo MMGD)
data = {
    "Ano": [2022, 2023, 2024],
    "Hidraulica (MW)": [109189, 108392, 108001],
    "Eolica (MW)": [23162, 28506, 32264],
    "Solar (MW)": [6917, 11212, 16075],
    "Termica (MW)": [39090, 39483, 37163],
    "Nuclear (MW)": [1990, 1990, 1990],
    "MMGD (MW)": [0, 26510, 34988],
}

# Criar DataFrame
df = pd.DataFrame(data)

# Calcular Média e Desvio Padrão
mean_values = [
    df['Hidraulica (MW)'].mean(),
    df['Eolica (MW)'].mean(),
    df['Solar (MW)'].mean(),
    df['Termica (MW)'].mean(),
    df['Nuclear (MW)'].mean(),
    df['MMGD (MW)'].mean(),
]

std_values = [
    df['Hidraulica (MW)'].std(),
    df['Eolica (MW)'].std(),
    df['Solar (MW)'].std(),
    df['Termica (MW)'].std(),
    df['Nuclear (MW)'].std(),
    df['MMGD (MW)'].std(),
]

labels = ['Hidráulica', 'Eólica', 'Solar', 'Térmica', 'Nuclear', 'MMGD']

# Criar gráfico de barras para Média e Desvio Padrão
plt.figure(figsize=(10, 6))
bar_width = 0.35
x = range(len(labels))

fig, ax = plt.subplots()

# Barras para Média (com uma cor padrão)
ax.bar(x, mean_values, bar_width, label='Média', color='royalblue')

# Barras para Desvio Padrão (deslocadas para a direita, com uma cor padrão)
ax.bar([i + bar_width for i in x], std_values, bar_width, label='Desvio Padrão', color='gray')

# Adicionar linhas de fundo para facilitar a visualização
ax.grid(True, axis='y', linestyle='--', alpha=0.7)

# Adicionar os rótulos
ax.set_xticks([i + bar_width / 2 for i in x])
ax.set_xticklabels(labels)
ax.set_title('Média e Desvio Padrão da Produção de Energia (MW) por Fonte de Energia')
ax.set_ylabel('Produção de Energia (MW)')
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de produção em MW para cada fonte de energia em 2022, 2023 e 2024
producoes = {
    'Hidráulica': [109189, 108392, 108001],
    'Eólica': [23162, 28506, 32264],
    'Solar': [6917, 11212, 16075],
    'Térmica': [39090, 39483, 37163],
    'Nuclear': [1990, 1990, 1990],
    'MMGD': [0, 26510, 34988]  # MMGD com 0 MW para 2022
}

# Preparando os dados para o boxplot
dados = list(producoes.values())

# Criando o gráfico de boxplot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotando o boxplot
ax.boxplot(dados, vert=True, patch_artist=True, 
           boxprops=dict(facecolor='skyblue', color='blue'),
           whiskerprops=dict(color='blue'),
           capprops=dict(color='blue'),
           flierprops=dict(marker='o', color='red', markersize=6))

ax.grid(True, axis='y', linestyle='--', color='gray', alpha=0.6)


# Definindo os rótulos
ax.set_xticklabels(producoes.keys(), fontsize=12)
ax.set_ylabel('Produção (MW)', fontsize=14)
ax.set_title('Produção Anual de Energia por Fonte', fontsize=16)

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```
